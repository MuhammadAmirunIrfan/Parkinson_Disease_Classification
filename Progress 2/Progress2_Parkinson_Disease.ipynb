## 4.5 Data Normalization/Scaling

### Why do we need scaling?

Different features have different ranges:
- `MDVP:Fo(Hz)` ranges from ~88 to ~260
- `MDVP:Jitter(%)` ranges from ~0.001 to ~0.03

Machine learning algorithms think bigger numbers are more important. Without scaling, the algorithm might focus too much on features with larger values.

### Standard Scaling (Z-score normalization)

**Formula:** `z = (x - mean) / standard_deviation`

**Result:** All features will have:
- Mean â‰ˆ 0
- Standard deviation â‰ˆ 1

````python
# Show why scaling is needed
print("Why We Need Scaling:")
print("="*60)
print("\nDifferent features have very different ranges:")
print(f"   - MDVP:Fo(Hz):     min={X['MDVP:Fo(Hz)'].min():.2f}, max={X['MDVP:Fo(Hz)'].max():.2f}")
print(f"   - MDVP:Jitter(%):  min={X['MDVP:Jitter(%)'].min():.5f}, max={X['MDVP:Jitter(%)'].max():.5f}")
print(f"   - HNR:             min={X['HNR'].min():.2f}, max={X['HNR'].max():.2f}")
print("\nâš ï¸ These different scales can bias ML algorithms!")
````

````python
# Import StandardScaler
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Apply Standard Scaling
print("Applying Standard Scaling:")
print("="*60)

# Create the scaler object
scaler = StandardScaler()

# Fit and transform the features
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for easier viewing
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("\nâœ“ Standard Scaling applied successfully!")
````

## 4.8 Checking for Outliers

Outliers are values that are significantly different from other values. We use the **IQR method** to detect them.

**IQR Method:**
- Q1 = 25th percentile
- Q3 = 75th percentile
- IQR = Q3 - Q1
- Outliers = values below Q1 - 1.5Ã—IQR or above Q3 + 1.5Ã—IQR

````python
# Function to count outliers
def count_outliers(data, column):
    """Count outliers using IQR method"""
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((data[column] < lower_bound) | (data[column] > upper_bound)).sum()
    return outliers

# Count outliers for each feature
print("Outlier Detection (IQR Method):")
print("="*60)

outlier_counts = {}
for col in X.columns:
    outliers = count_outliers(X, col)
    outlier_counts[col] = outliers

# Display features with outliers
print("\nFeatures with outliers:")
for col, count in outlier_counts.items():
    if count > 0:
        print(f"   - {col}: {count} outliers")

print(f"\nNote: We keep outliers in medical data as they may contain important information!")
````

## 4.9 Binning 

Binning is the process of grouping continuous values into discrete bins or categories.

**Example:** Converting age (continuous) into age groups (categorical)
- 0-18 â†’ "Young"
- 19-60 â†’ "Adult"
- 60+ â†’ "Senior"



````python
# Example of binning: Convert HNR into categories
print("Example of Binning:")
print("="*60)

# Create a copy of the cleaned dataframe
df_binned = df_clean.copy()

# Define bins for HNR
bins = [0, 15, 20, 25, 40]
labels = ['Low', 'Medium', 'High', 'Very High']

# Apply binning
df_binned['HNR_category'] = pd.cut(df_binned['HNR'], bins=bins, labels=labels)

print("\nOriginal HNR values (first 10):")
print(df_clean['HNR'].head(10).tolist())

print("\nBinned HNR categories (first 10):")
print(df_binned['HNR_category'].head(10).tolist())

print("\nDistribution of HNR categories:")
print(df_binned['HNR_category'].value_counts())
````

## 4.10 Indicator Variables (One-Hot Encoding)

Indicator variables (dummy variables) convert categorical data into binary (0/1) columns.

**Example:** If we have a 'Color' column with values ['Red', 'Blue', 'Green']:
```
Color    â†’   Color_Red  Color_Blue  Color_Green
Red      â†’      1          0           0
Blue     â†’      0          1           0
Green    â†’      0          0           1
```

````python
# Example of creating indicator variables from binned data
print("Example of Indicator Variables (One-Hot Encoding):")
print("="*60)

# Create indicator variables from HNR_category
indicator_vars = pd.get_dummies(df_binned['HNR_category'], prefix='HNR')

print("\nOriginal categorical column:")
print(df_binned['HNR_category'].head(10).tolist())

print("\nIndicator variables (first 10 rows):")
print(indicator_vars.head(10))

print("\n Note: Each category becomes a separate column with 0 or 1 values")
````
